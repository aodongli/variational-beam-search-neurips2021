```python
def is_connected(x, y):
    for x0, y0, x1, y1 in zip(x[:-1], y[:-1], x[1:], y[1:]):
        if np.abs(x0-x1) > 3 or np.abs(y0-y1) > 3:
            # check disconnectedness if the neighbouring position is larger than 3 ft.
            return False
    return True

def sample_movement_from_games(game_json_list, rng):
    '''
    Returns:
    The first returning argument is a position set used for the tracking model [[x_t1, y_t2], ...];
    The second returning argument is a combo of the original coordinate set and the game clock set [x, y, t].
    '''
    import glob
    pos_set = []
    ori_movements = []
    for i, g in enumerate(game_json_list):
        json_file = glob.glob('data/'+g+'.json/*.json')[0]
        game = Game(game_json_file=json_file)
        print('Game:', i)
        print(game)
        names = game.get_player_list()
        x = []
        while not x:
            # repeat until get a player with movement information
            name = rng.choice(names, replace=False)
            event_id = rng.choice(game.num_events, replace=False)
            x, y, t = game.get_player_movement(name, event_id)
            if not is_connected(x, y):
                x = []
        pos = np.array([[x_1, x_2] for x_1, x_2 in zip(x, y)])
        print('Sampled player, event id, and shape:', name, event_id, pos.shape)
        pos_set.append(pos)
        ori_movements.append([x, y, t])
    return pos_set, ori_movements
```
